```python
# Import all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings — show more columns/rows when printing dataframes
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 150)

print("Libraries imported successfully")
```

```python
# Load the attainment dataset to take a first look
attainment = pd.read_csv('../data/attainment8.csv', low_memory=False)

print("Shape:", attainment.shape)
print("\nColumn names (first 20):")
print(attainment.columns.tolist()[:20])
```

```python
# What years and geographic levels does this dataset cover
print("Years available:", sorted(attainment['time_period'].unique()))
print("\nGeographic levels available:", attainment['geographic_level'].unique())
```

```python
# Keep only Local Authority District level rows
la_attainment = attainment[attainment['geographic_level'] == 'Local authority district'].copy()

print("Shape after filtering to LA level:", la_attainment.shape)
print("Number of unique local authorities:", la_attainment['lad_name'].nunique())
```

```python
# Understand how the data is broken down
print("breakdown_topic values:", la_attainment['breakdown_topic'].unique())
print("\ndisadvantage_status values:", la_attainment['disadvantage_status'].unique())
```

```python
# Look at the key column we care about most
print(la_attainment[['time_period', 'lad_name', 'disadvantage_status', 'attainment8_average', 'pupil_count']].head(10))
```
