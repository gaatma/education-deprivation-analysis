## Section 3 — Data Cleaning and Preprocessing

This section handles missing values, standardises column names, aggregates the IMD data from LSOA to Local Authority level, and merges all three datasets into one analysis-ready table. Before and after statistics are recorded at each step to document the impact of cleaning.

```python
#Custom function to clean the attainment dataset
def clean_attainment_data(df):
    """
    Clean the DfE Attainment 8 dataset.

    Steps performed:
    - Filters to Local Authority District level only
    - Removes duplicate rows
    - Converts attainment8_average to numeric (coercing any bad values to NaN)

    Args:
        df (pd.DataFrame): Raw attainment dataframe

    Returns:
        pd.DataFrame: Cleaned dataframe ready for merging
    """
    df = df.copy()
    df = df[df['geographic_level'] == 'Local authority district']
    df = df.drop_duplicates()
    df['attainment8_average'] = pd.to_numeric(df['attainment8_average'], errors='coerce')
    return df

# BEFORE stats
print("BEFORE cleaning — shape:", attainment.shape)

# Apply the cleaning function
la_attainment_clean = clean_attainment_data(attainment)

# AFTER stats
print("AFTER cleaning — shape:", la_attainment_clean.shape)
print("Nulls in attainment8_average after cleaning:", la_attainment_clean['attainment8_average'].isnull().sum())
```

```python
# Standardise column names across the two IMD files
# This makes both files consistent so we can combine them later

def standardise_imd_columns(df, year):
    """
    Standardise column names for an IMD dataset and tag it with its year.

    Args:
        df (pd.DataFrame): Raw IMD dataframe (2019 or 2025)
        year (int): The year this IMD data represents, e.g. 2019 or 2025

    Returns:
        pd.DataFrame: Dataframe with standardised column names and a 'dep_year' column
    """
    df = df.copy()
    df = df.rename(columns={
        df.columns[2]: 'lad_code',     # 3rd column is always the LA code
        df.columns[3]: 'lad_name',     # 4th column is always the LA name
        'Index of Multiple Deprivation (IMD) Score': 'imd_score',
        'Index of Multiple Deprivation (IMD) Rank (where 1 is most deprived)': 'imd_rank',
        'Index of Multiple Deprivation (IMD) Decile (where 1 is most deprived 10% of LSOAs)': 'imd_decile',
        'Income Score (rate)': 'income_score',
        'Employment Score (rate)': 'employment_score',
        'Education, Skills and Training Score': 'education_score',
        'Health Deprivation and Disability Score': 'health_score',
        'Crime Score': 'crime_score',
    })
    df['dep_year'] = year
    return df

imd2019_clean = standardise_imd_columns(imd2019, 2019)
imd2025_clean = standardise_imd_columns(imd2025, 2025)

print("IMD 2019 columns now:", imd2019_clean.columns.tolist()[:10])
```

```python
# Aggregate from LSOA level to Local Authority level
# We take the MEAN of all LSOA scores within each local authority
# This gives us one representative deprivation score per LA

def aggregate_to_la_level(df):
    """
    Aggregate LSOA-level IMD data up to Local Authority District level
    by taking the mean of each score across all LSOAs in that LA.

    Args:
        df (pd.DataFrame): LSOA-level IMD dataframe with lad_code and lad_name

    Returns:
        pd.DataFrame: One row per local authority with averaged scores
    """
    agg_cols = ['imd_score', 'imd_rank', 'imd_decile', 'income_score',
                'employment_score', 'education_score', 'health_score', 'crime_score']

    la_level = df.groupby(['lad_code', 'lad_name', 'dep_year'])[agg_cols].mean().reset_index()
    return la_level

imd2019_la = aggregate_to_la_level(imd2019_clean)
imd2025_la = aggregate_to_la_level(imd2025_clean)

print("IMD 2019 aggregated to LA level — shape:", imd2019_la.shape)
print(imd2019_la.head())
```

```python
# Stack the 2019 and 2025 deprivation data into one table
# This lets us compare deprivation over time for the same local authorities

deprivation_combined = pd.concat([imd2019_la, imd2025_la], ignore_index=True)

print("Combined deprivation data shape:", deprivation_combined.shape)
print("\nYears present:", deprivation_combined['dep_year'].unique())
print("Unique local authorities:", deprivation_combined['lad_code'].nunique())
```

```python
# Merge attainment data with deprivation data on local authority code
# We merge on lad_code since names can sometimes differ slightly in formatting

# First, let's check how many LA codes actually overlap
attainment_codes = set(la_attainment_clean['lad_code'].unique())
deprivation_codes = set(deprivation_combined['lad_code'].unique())

print("Local authorities in attainment data:", len(attainment_codes))
print("Local authorities in deprivation data:", len(deprivation_codes))
print("Overlapping local authorities:", len(attainment_codes & deprivation_codes))
print("In attainment but NOT in deprivation data:", len(attainment_codes - deprivation_codes))
```

```python
# merging
merged_df = la_attainment_clean.merge(
    deprivation_combined,
    left_on='lad_code',
    right_on='lad_code',
    how='left'
)

print("\nMerged dataframe shape:", merged_df.shape)
print("Rows where deprivation data is missing:", merged_df['imd_score'].isnull().sum())
```

```python
# Map each academic year to the nearest deprivation snapshot year
# Years 2018/19 to ~2021/22 are closer to the 2019 IMD release
# Years 2022/23 onwards are closer to the 2025 IMD release

def map_to_deprivation_year(time_period):
    """
    Map an academic year (e.g. 202122) to the closest available
    deprivation data year (2019 or 2025).

    Args:
        time_period (int): Academic year code, e.g. 201819, 202425

    Returns:
        int: 2019 or 2025 — whichever deprivation snapshot is closest
    """
    year = int(str(time_period)[:4])  # take the first 4 digits as the starting year
    if year <= 2021:
        return 2019
    else:
        return 2025

la_attainment_clean['matched_dep_year'] = la_attainment_clean['time_period'].apply(map_to_deprivation_year)

print(la_attainment_clean[['time_period', 'matched_dep_year']].drop_duplicates())
```

```python
# redo the merge using BOTH lad_code AND matched year
merged_df = la_attainment_clean.merge(
    deprivation_combined,
    left_on=['lad_code', 'matched_dep_year'],
    right_on=['lad_code', 'dep_year'],
    how='left'
)

print("Merged dataframe shape:", merged_df.shape)
print("Rows where deprivation data is missing:", merged_df['imd_score'].isnull().sum())
print("\nPercentage matched:", round((1 - merged_df['imd_score'].isnull().sum() / len(merged_df)) * 100, 1), "%")
```

```python
# Clean up duplicate column names after merging
# lad_name_x = name from attainment data, lad_name_y = name from deprivation data
# They should be identical for matched rows, so we keep one and drop the other

merged_df = merged_df.rename(columns={'lad_name_x': 'lad_name'})
merged_df = merged_df.drop(columns=['lad_name_y'])

print("Columns now:", [col for col in merged_df.columns if 'name' in col.lower()])
```

```python
# Identify exactly which local authorities did not match


unmatched = merged_df[merged_df['imd_score'].isnull()]['lad_name'].unique()
print("Local authorities with no deprivation data match:")
for la in sorted(unmatched):
    print(" -", la)

print(f"\nTotal unmatched local authorities: {len(unmatched)}")
print(f"Total unmatched rows: {merged_df['imd_score'].isnull().sum()} out of {len(merged_df)}")
```

```python
# Handle unmatched rows
# Decision: these local authorities cannot be analysed without deprivation data,
# so we remove them. This represents only 1.7% of rows (360 out of 21,611) —
# a small, well-documented loss that does not bias the wider analysis.

print("BEFORE removing unmatched rows:", merged_df.shape)

merged_df_clean = merged_df.dropna(subset=['imd_score']).copy()

print("AFTER removing unmatched rows:", merged_df_clean.shape)
print(f"\nRows removed: {merged_df.shape[0] - merged_df_clean.shape[0]} ({merged_df['imd_score'].isnull().sum()} were missing)")
```

### Handling Unmatched Local Authorities

15 local authorities present in the attainment dataset could not be matched to deprivation data, resulting in 360 unmatched rows (1.7% of the merged dataset). Investigation showed these are primarily local authorities affected by boundary reorganisations between 2019 and 2025 (e.g. Northamptonshire's split into North and West Northamptonshire in 2021, and Buckinghamshire's transition to a single unitary authority).

**Decision:** These 360 rows were removed from the analysis dataset, as deprivation data cannot be reliably imputed for areas with no equivalent geography in either IMD dataset. This represents a small, well-documented loss (1.7% of rows) that does not meaningfully bias the wider analysis of 324 local authorities across England.

```python
# Use NumPy to catch any infinite or extreme values
# Flow-based and rate-based statistics can sometimes produce inf values
# when a denominator is zero — this is standard practice to check for

inf_count = np.isinf(merged_df_clean.select_dtypes(include=[np.number])).sum().sum()
print("Infinite values found:", inf_count)

# Replace any that exist with NaN, then drop them
merged_df_clean = merged_df_clean.replace([np.inf, -np.inf], np.nan)

# Calculate overall data quality using NumPy
total_cells = merged_df_clean.shape[0] * merged_df_clean.shape[1]
null_cells = merged_df_clean.isnull().sum().sum()
data_quality_rate = np.round((1 - null_cells / total_cells) * 100, 2)

print(f"Total cells: {total_cells}")
print(f"Null cells: {null_cells}")
print(f"Overall data completeness: {data_quality_rate}%")
```

```python
# Use NumPy to flag statistical outliers in Attainment 8 scores
# A common approach: anything more than 3 standard deviations from the mean

mean_a8 = np.mean(merged_df_clean['attainment8_average'])
std_a8 = np.std(merged_df_clean['attainment8_average'])

lower_bound = mean_a8 - (3 * std_a8)
upper_bound = mean_a8 + (3 * std_a8)

print(f"Mean Attainment 8: {np.round(mean_a8, 2)}")
print(f"Standard deviation: {np.round(std_a8, 2)}")
print(f"Outlier bounds: below {np.round(lower_bound,2)} or above {np.round(upper_bound,2)}")

outliers = merged_df_clean[
    (merged_df_clean['attainment8_average'] < lower_bound) |
    (merged_df_clean['attainment8_average'] > upper_bound)
]

print(f"\nNumber of statistical outliers found: {len(outliers)}")
print(outliers[['lad_name', 'time_period', 'disadvantage_status', 'attainment8_average', 'pupil_count']])
```

```python
# Investigate small pupil-count instability
# The Isles of Scilly has extremely small pupil cohorts, causing the
# average Attainment 8 score to swing wildly year to year (19.0 to 70.0)
# This is a sample-size problem, not a deprivation signal

print("Pupil count distribution (smallest 15):")
print(merged_df_clean[['lad_name', 'time_period', 'disadvantage_status', 'pupil_count']]
      .sort_values('pupil_count')
      .head(15))
```

```python
# set a minimum pupil count threshold for inclusion in analysis
# very small cohorts because individual averages are not statistically meaningful

MIN_PUPIL_COUNT = 5

before_count = len(merged_df_clean)
merged_df_clean = merged_df_clean[merged_df_clean['pupil_count'] >= MIN_PUPIL_COUNT].copy()
after_count = len(merged_df_clean)

print(f"BEFORE pupil-count filter: {before_count} rows")
print(f"AFTER pupil-count filter: {after_count} rows")
print(f"Rows removed: {before_count - after_count} ({np.round((before_count-after_count)/before_count*100, 2)}%)")
```

### Outlier Investigation — Small Cohort Instability

Statistical outlier detection (±3 standard deviations) flagged 4 rows, all belonging to the **Isles of Scilly** with `pupil_count = 1`. Investigation showed these are not data errors but a genuine small-number statistics problem: with only one disadvantaged pupil sitting exams in a given year, that single pupil's score becomes "the average," producing unstable swings (19.0 to 70.0) unrelated to actual deprivation effects.

**Decision:** A minimum pupil count threshold of 5 was applied across the dataset, consistent with DfE's own suppression practices for small cohorts. This removed [X]% of rows while preserving the integrity of local-authority-level averages used in correlation and modelling analysis.

This demonstrates the difference between a genuine outlier (which should be retained, e.g. a real local authority that significantly over- or under-performs) and an unstable statistic caused by an insufficient sample size (which should be excluded or treated with caution).

```python
#Confirm the outlier issue is resolved

mean_a8 = np.mean(merged_df_clean['attainment8_average'])
std_a8 = np.std(merged_df_clean['attainment8_average'])
lower_bound = mean_a8 - (3 * std_a8)
upper_bound = mean_a8 + (3 * std_a8)

outliers_after = merged_df_clean[
    (merged_df_clean['attainment8_average'] < lower_bound) |
    (merged_df_clean['attainment8_average'] > upper_bound)
]

print(f"Outliers remaining after pupil-count filter: {len(outliers_after)}")
if len(outliers_after) > 0:
    print(outliers_after[['lad_name', 'time_period', 'disadvantage_status', 'attainment8_average', 'pupil_count']])
```

### Outlier Investigation — Small Cohort Instability

Statistical outlier detection (±3 standard deviations) flagged 4 rows, all belonging to the **Isles of Scilly** and **City of London**, both with `pupil_count = 1`. Investigation showed these are not data errors but a genuine small-number statistics problem: with only one pupil sitting exams in a given year and disadvantage category, that single pupil's score becomes "the average," producing unstable swings (e.g. 19.0 to 70.0) unrelated to any real deprivation effect.

**Decision:** A minimum pupil count threshold of 5 was applied across the dataset, consistent with DfE's own suppression practices for small cohorts. This removed 31 rows (0.15% of the dataset) and resolved all detected outliers — confirming the issue was driven entirely by small-sample instability in England's two smallest local authorities, the Isles of Scilly and the City of London, rather than by deprivation patterns.

## Section 4 — In-Depth Analysis with Statistical Insights

This section answers our four research questions through statistical analysis:

1. Do more deprived local authorities consistently produce lower Attainment 8 scores? *(Correlation analysis)*
2. Has the attainment gap between disadvantaged and non-disadvantaged pupils changed between 2018/19 and 2024/25? *(Gap analysis over time)*
3. Which local authorities perform above or below expectations relative to their deprivation level? *(Residual analysis)*
4. Can educational attainment be predicted using deprivation-related indicators? *(Predictive modelling — Section 6)*

```python
# Research Question 1: Does deprivation correlate with Attainment 8?
# We use Pearson correlation — appropriate for two continuous variables

from scipy.stats import pearsonr

# Use the most recent year, Total population (not split by disadvantage)
latest_total = merged_df_clean[
    (merged_df_clean['time_period'] == 202425) &
    (merged_df_clean['disadvantage_status'] == 'Total')
].copy()

# Calculate correlation between IMD score and Attainment 8
correlation, p_value = pearsonr(latest_total['imd_score'],
                                 latest_total['attainment8_average'])

print(f"Pearson correlation coefficient: {correlation:.4f}")
print(f"P-value: {p_value:.6f}")

# Interpret the result
if p_value < 0.05:
    significance = "statistically significant"
else:
    significance = "not statistically significant"

if abs(correlation) > 0.7:
    strength = "strong"
elif abs(correlation) > 0.4:
    strength = "moderate"
else:
    strength = "weak"

direction = "negative" if correlation < 0 else "positive"

print(f"\nInterpretation: {strength} {direction} correlation, {significance} (p < 0.05)")
print(f"\nThis means: as deprivation score increases, Attainment 8 {'decreases' if correlation < 0 else 'increases'}.")
```
